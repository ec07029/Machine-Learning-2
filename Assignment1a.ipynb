```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data
df = pd.read_csv("Advertising.csv")
df.iloc[:5, [1, -1]]
```

```python
# Calculate minimum using minimize() function
from scipy.optimize import minimize

def obj(w):
    m, b = w
    x = df.iloc[:5, 1].values
    y = df.iloc[:5, -1].values
    pred = m * x + b

    return ((pred - y) ** 2).mean()

res = minimize(obj, x0 = [0, 0])
res.x
```

```python
# Calculate minimum with normal equation
y = pd.DataFrame(df.iloc[:5, -1]) # Sales data (output)
y
```

```python
X = pd.DataFrame(df.iloc[:5, 1]) # TV data (input)
one = [1, 1, 1, 1, 1] # required for matrix
X["one"] = one
X
```

```python
X_transpose = np.transpose(X) # TV data transposed
X_transpose
```

```python
X_transpose_X = np.dot(X_transpose, X) # Dot product of TV data transposed x TV data
X_transpose_X
```

```python
X_transpose_y = np.dot(X_transpose, y) # Dot product of TV data transposed x Sales data
X_transpose_y
```

```python
X_transpose_X_inv = np.linalg.inv(X_transpose_X) # inverse of (TV transposed data x TV data)
```

```python
beta = np.dot( X_transpose_X_inv, X_transpose_y ) # Normal equation
beta
```

```python
# Output from normal equation matches output from minimize() function
print("Minimize function: ", res.x)
print("Normal equation: ", beta)
```
